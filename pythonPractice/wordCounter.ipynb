```python
words = []
with open('wordCounter.txt', 'r') as textFile:
    for line in textFile:
        words.extend(line.split())
```

```python
from collections import Counter
```

```python
import pandas as pd

pd.Series(Counter(words)).sort_values(ascending = False).head(20)
```

## Adding stopwords to remove

```python
words
```

```python
# creating list of words to remove
removeList = ['a', 'an', 'and', 'the', 'is', 'of', 'to', 'for', 'in', 'on', 'am', 'are', 'as', 'at'
             'be', 'by', 'has', 'had', 'i', 'it']
```

```python
filteredList = [word for word in words if not word.lower() in removeList]
filteredList
```

```python
pd.Series(Counter(filteredList)).sort_values(ascending = False)
```
